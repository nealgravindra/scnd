```python
def kendall(AnnData, n=100, save=None, verbose=False):
    if verbose:
        tic = time.time()
    results = pd.DataFrame(columns=['tau', 'P'], index=AnnData.var_names.to_list())
    cell_sample = random.sample(AnnData.obs.index.to_list(), n)   
    for i, g in enumerate(AnnData.var_names):
        x = np.asarray(AnnData[AnnData.obs.index.isin(cell_sample), g].X.todense()).flatten()
        ximp = np.asarray(AnnData[AnnData.obs.index.isin(cell_sample), g].layers['imputed']).flatten()
        tau, p = kendalltau(x, ximp)
        results.loc[g, 'tau'] = tau
        results.loc[g, 'P'] = p
        
        if i % 1000 == 0 and i != 0 and verbose:
            print('Through {:.2f}-% in {:.1f}-s. Iterative saved'.format(100*(i/results.shape[0]), time.time() - tic))
            if save is not None:
                results.to_csv(save)      
    if save is not None:
        results.to_csv(save)
    return results
```

```python
def check_corrs(AnnData, gene_list, obs_vars, 
                n=100, save=None, verbose=False):
    if verbose:
        tic = time.time()
    results = pd.DataFrame(columns=[i+'_tau' for i in obs_vars]+[i+'_P' for i in obs_vars], 
                           index=gene_list)
    cell_sample = random.sample(AnnData.obs.index.to_list(), n)   
    for i, g in enumerate(gene_list):
        x = np.asarray(AnnData[AnnData.obs.index.isin(cell_sample), g].X.todense()).flatten()
        for j in obs_vars:
            y = AnnData[AnnData.obs.index.isin(cell_sample), :].obs[j].to_numpy()
            tau, p = kendalltau(x, y)
            results.loc[g, j+'_tau'] = tau
            results.loc[g, j+'_P'] = p 
        if i % 25 == 0 and i != 0 and verbose:
            print('  through {:.2f}-% in {:.1f}-s'.format(100*(i/results.shape[0]), time.time() - tic))
            if save is not None:
                results.to_csv(save)
                print('    itereatively saved.')
    if save is not None:
        results.to_csv(save)
    return results
```

```python
%%memit 

tic = time.time()
from scipy.stats import spearmanr
rho, p = np.zeros((wt.shape[1],)), np.zeros((wt.shape[1],)) 
X = np.asarray(wt.X.todense()) # memory intensive
for i in range(wt.shape[1]):
    # check where that gene is >0 counts
    nonzero_idx = np.asarray((wt.raw[:, i].X>0).todense()).flatten()
    if np.sum(nonzero_idx) == 0: 
        #continue; else:
        print('No counts for {}'.format(wt.var_names[i]))
    rho[i], p[i] = spearmanr(X[nonzero_idx, i], wt.layers['imputed'][nonzero_idx, i], nan_policy='raise')
    if i % 200 == 0 and i!=0:
        print('{:.2f}-s elapsed\tthrough {:.2f}-% of iterations'.format(time.time() - tic, 100*(i/wt.shape[1])))
toc = time.time()
    
print('Correlations in {:.2f}-min'.format((toc-tic)/60))
```

```python
results_wt = pd.DataFrame({'rho':rho, 'P':p}, index=wt.var_names.to_list())
results_wt.to_csv(os.path.join(pfp, 'imp_corr_wt_nonzero.csv'))
```
